## Line Chart Plotting

````python
from random import choice
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])
````

````python
plt.plot(x, y,c='red')
plt.plot(y,c='blue')
plt.show()
````

#### fmt(format string) argument
#### Syntax:
1. Color (optional)
2. Marker (optional)
3. LineStyle (optional)

````python
plt.plot(x, y, 'ro-')
plt.show()
````

#### Basic Arguments
- x, y: Coordinates for the data points to plot.
- fmt: A format string, like '-', 'o', or '-o', to specify line and marker styles.
#### Line Properties
- color or c: Specify the color of the line, e.g., 'blue', 'r', '#FFA500'.
- linestyle or ls: Change the style of the line, e.g., '-' for solid, '--' for dashed.
- linewidth or lw: Set the width of the line, e.g., 2.5.
#### Marker Properties
- marker: Choose the marker style, e.g., 'o', 'x', '*', etc.
- markersize or ms: Size of the markers, e.g., 10.
- markerfacecolor or mfc: Set the fill color of markers.
- markeredgewidth or mew: Width of the marker edge.
- markeredgecolor or mec: Color of the marker edge.
#### Additional Style Arguments
- alpha: Set the transparency of the plot, ranging from 0 (transparent) to 1 (opaque).
- label: Label for the legend.
- drawstyle: Change how points are connected ('default', 'steps', 'steps-pre', 'steps-mid', 'steps-post').
#### Data Limits and Scaling
- scalex: Auto-scale the x-axis, default is True.
- scaley: Auto-scale the y-axis, default is True.
#### Other Arguments
- antialiased or aa: Enable or disable antialiasing for the line.
- dash_capstyle: Style of the end of dashes, e.g., 'butt', 'round', 'projecting'.
- solid_capstyle: Style of the end of solid lines, e.g., 'butt', 'round', 'projecting'.
- solid_joinstyle: Join style for solid lines, e.g., 'miter', 'round', 'bevel'.
- dash_joinstyle: Join style for dashed lines, e.g., 'miter', 'round', 'bevel'.
- dash_offset: Offset for dashes.
- dashes: A list of dashes for custom dash patterns.

````python
plt.plot(
    x, y, 
    color='blue',           # (c)
    linestyle= choice([choice(['-', '--', '-.', ':', '']), choice(['solid', 'dashed', 'dashdot', 'dotted', 'none'])]),         # (ls)
    linewidth=2.5,          # (lw)
    marker= choice(['.', ',', 'o', 'v', '^', '<', '>', '1', '2', '3', '4', 's', 'p', '*', 'h', 'H', '+', 'x', 'D', 'd', '|', '_', 'P', 'X', 0, 1, 2, 3, 'None', '',' ','$\u266B$']),  # Circle marker
    markersize=8,           # (ms)
    markerfacecolor='red',  # (mfc)
    markeredgewidth=4,      # (mew)
    markeredgecolor='black',# (mec)
    alpha=0.75,             # Transparency
    label='MyLine',         # Label for legend
    scalex=True,            # Autoscale x-axis
    scaley=True,            # Autoscale y-axis
    antialiased=True,       # (aa) Enable antialiasing
    dash_capstyle='round',  # Style of end of dashes
    solid_capstyle='projecting', # Style of solid line ends
    #dashes=[5, 2, 10, 2]    # Custom dash pattern: 5 on, 2 off, 10 on, 2 off
)
del choice
plt.show()
````

#### Labeling

````python
plt.plot(x, y,c='red')

font_properties = {
    'fontsize': 14,
    'fontweight': 'bold',
    'color': 'darkred',
    'fontstyle': 'italic',
    'fontfamily': 'serif',
    'horizontalalignment': 'center',
    'verticalalignment': 'bottom',
    'rotation': 0
}

plt.xlabel(
    'xlabel',      # str: The label text
    labelpad=2.3,    # float: Spacing in points from the axis bounding box
    loc='center',         # str: 'left', 'center', or 'right' for xlabel; 'bottom', 'center', or 'top' for ylabel
    #fontdict=font_properties,    # dict: A dictionary to override the default text properties (size, family, weight, etc.)
    fontsize=14,    # float or str: Size of the font (e.g., 'large', 'medium', 'small', etc.)
    color='red',       # color: Label text color
    #'horizontalalignment', # str: Horizontal alignment ('left', 'center', 'right')
    #'verticalalignment',   # str: Vertical alignment ('top', 'center', 'bottom')
    fontweight='bold',  # str: Text weight (e.g., 'bold', 'normal')
    rotation=30,    # float or {'vertical', 'horizontal'}: Rotation angle of the label
    family='serif',      # str: Font family (e.g., 'serif', 'sans-serif', 'monospace')
    #'label',       # str: Label for the legend
    #'x',           # float: X position for xlabel (usually set automatically)
    #'y'            # float: Y position for ylabel (usually set automatically)
)

plt.ylabel("This is ylabel")
````

#### Grid

````python
plt.grid(
    True,           # Show grid
    which= choice(['major', 'minor', 'both']),     # Apply to both major and minor grid lines
    axis=choice(['x', 'y', 'both']),         # Default is 'both'
    color='hotpink',     # Set grid line color
    linestyle='--',   # Set grid line style
    linewidth=2,    # Set grid line width
    alpha=0.8         # Set transparency
)

plt.show()
````
